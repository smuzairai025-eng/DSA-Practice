```python
class Library():
    __book_list=["physics","math","Bio","chemistry","english","urdu"]
    __borrowed_data={}

    def __init__(self):
        # self.name=input("Enter your name")
        # self.total_book_list=["physics","math","Bio","chemistry","english","urdu"]
        self.__name=input("Enter your name")
        self.__borrowed_books=[]

        self.__menu()

    def __menu(self):
        while True:
            user_input=input("""Choose an option
                                1.choose 1 to display books
                                2.choosw 2 to borrow a book
                                3.choose 3 to return a book
                                4.choose 4 to add a new book
                                5.choose 5 to see borrowed data
                                6.choose 6 to exit""")
            
            if user_input=="1":
                self.display_books()
            elif user_input=="2":
                self.borrow_book()
            elif user_input=="3":
                self.return_book()
            elif user_input=="4":
                self.add_book()
            elif user_input=="5":
                self.total_borrowed_data()
            elif user_input=="6":
                break
            else:
                print("choose valid option")
    
    def display_books(self):
        print(f"Available Books :{Library.__book_list}")

    
    def borrow_book(self):
        book=input("Enter book name")
        if book in Library.__book_list:
            Library.__book_list.remove(book)
            self.__borrowed_books.append(book)
            Library.__borrowed_data.update({self.__name : book})
            print(f"{self.__borrowed_books} Borrowed")
            
        elif book in Library.__borrowed_data.values():
            print(f"{book} already borrowed")
        else:
            print(f"{book} not available")

    def return_book(self):
        if self.__borrowed_books is not None:
            book=input("which book do you want to return")
            if book in self.__borrowed_books:
                Library.__book_list.append(book)
                self.__borrowed_books.remove(book)
                if self.__name in Library.__borrowed_data and Library.__borrowed_data[self.__name]==book:
                    del Library.__borrowed_data[self.__name]
                print(f"{book} Book returned")
                
                print(f"Book list :{self.__book_list}, Borrowed books : {self.__borrowed_books}")
            else:
                print(f"you do not have borrowed {book} book")
        else:
                print("you do not have a book")



    def add_book(self):
        user_input=input("which book do you want to add to collection")
        if user_input not in Library.__book_list and user_input not in Library.__borrowed_data.values():
            Library.__book_list.append(user_input)
            
        else:
            print(f"{user_input} already available in collection")

    def total_borrowed_data(self):
        print(f"Borrowed data is {Library.__borrowed_data}")


```

```python
def bubble_sort(arr):

    for i in range (len(arr)-1):
        f=0
        for j in range (len(arr)-1-i):
            if arr[j]>arr[j+1]:
                temp=arr[j]
                arr[j]=arr[j+1]
                arr[j+1]=temp
                f=1
        if f==0:
            print("sorted array")
            return arr
    print("complete")
    return arr
```

```python
arr=[1,2,3,4,7,5]
bubble_sort(arr)
```

```python
def selection_sort(arr):
    for i in range(len(arr)-1):
        min=i
        for j in range(i+1,len(arr)):
            if arr[j]<arr[min]:
                min=j
        temp=arr[i]
        arr[i]=arr[min]
        arr[min]=temp
                
    return arr
```

```python
arr=[3,2,4,1,5,7,6]
selection_sort(arr)
```

```python
def merge_sorted(arr1,arr2):

    i=0
    j=0
    merged=[]

    while i < len(arr1) and j < len(arr2):
        if arr1[i]<arr2[j]:
            merged.append(arr1[i])
            i+=1
        else:
            merged.append(arr2[j])
            j+=1
    while i < len(arr1):
        merged.append(arr1[i])
        i+=1
    while j < len(arr2):
        merged.append(arr2[j])
        j+=1
    return merged
```

```python
arr1=[1,3,5,7]
arr2=[2,4,6,8,9,10,12,13,14,15]
merge_sorted(arr1,arr2)
```

```python
def merge_sort(arr):

    if len(arr)==1:
        return arr

    mid=len(arr)//2

    left=arr[:mid]
    right=arr[mid:]

    left_arr=merge_sort(left)
    right_arr=merge_sort(right)

    sorted=merge_sorted(left_arr,right_arr)

    return sorted
    
```

```python
merge_sort([6,4,2,7,1,9])
```

```python
def bubble_sort(arr):

    for i in range (len(arr)-1):
        f=0
        for j in range (len(arr)-1-i):
            if arr[j]>arr[j+1]:
                temp=arr[j]
                arr[j]=arr[j+1]
                arr[j+1]=temp
                f=1
        if f==0:
            print("sorted array")
            return arr
    print("complete")
    return arr
```

```python
arr=[1,2,3,4,7,5]
bubble_sort(arr)
```

```python
def selection_sort(arr):
    for i in range(len(arr)-1):
        min=arr[i]
        for j in range(i+1,len(arr)):
            if arr[j]<min:
                temp=arr[i]
                arr[i]=arr[j]
                arr[j]=temp
                
    return arr
```

```python
arr=[3,2,4,1,5,7,6]
selection_sort(arr)
```

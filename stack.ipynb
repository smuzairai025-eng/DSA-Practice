```python
class Node:
    def __init__(self,value):
        self.data=value
        self.next=None

class Stack:
    def __init__(self):
        self.top=None
        self.n=0
    
    def is_empty(self):
        return self.top==None
    
    def push(self,value):
        new_node=Node(value)
        new_node.next=self.top
        self.top=new_node
        self.n+=1

    def traverse(self):
        temp=self.top
        res=''

        while temp!=None:
            data=temp.data
            temp=temp.next
            print(data)
        #     res=res+str(data)
        # return res
    
    def pop(self):
        if self.is_empty():
            return "empty stack"
        value=self.top.data
        self.top=self.top.next
        self.n-=1
        return value
    def peek(self):
        a=self.top.data
        return a

    def size(self):
        return self.n
```

```python
#Reverse the string using stack
def reverse(text):
    s=Stack()
    for i in text:
        s.push(i)
    res=''
    while not s.is_empty():
        res=res+str(s.pop())
    return res
        

    
```

```python
#Text Editin problem Undo-Redo
def edit_str(text,pattern):
    u=Stack()
    r=Stack()

    for i in text:
        u.push(i)
    for j in pattern:
        if j == 'u':
            val=u.pop()
            r.push(val)
        else:
            v=r.pop()
            u.push(v)
    res=''
    while not u.is_empty():
        data=u.pop()
        res=str(data)+res
    return res
```

```python
text='uzair'
pattern=''
edit_str(text,pattern)
```

```python
L=[
        [0,0,1,1],
        [1,0,1,0],
        [0,0,1,0],
        [0,1,1,0]
    ]
find_celeb(L)
```

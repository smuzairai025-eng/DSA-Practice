```python
class Node:
    def __init__(self,value):
        self.data=value
        self.next=None

class Stack:
    def __init__(self):
        self.top=None
        self.n=0
    
    def is_empty(self):
        return self.top==None
    
    def push(self,value):
        new_node=Node(value)
        new_node.next=self.top
        self.top=new_node
        self.n+=1

    def traverse(self):
        temp=self.top

        while temp!=None:
            print(temp.data)
            temp=temp.next
    
    def pop(self):
        if self.is_empty():
            return "empty stack"
        self.top=self.top.next
        self.n-=1

    def size(self):
        return self.n
```

```python
s=Stack()
```

```python
s.push(4)
s.push(3)
s.push(2)
s.push(1)
```

```python
s.size()
```

```python
s.traverse()
```

```python
s.pop()
```

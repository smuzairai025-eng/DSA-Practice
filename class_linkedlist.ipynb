```python
class Node():

    def __init__(self,value):
        self.data=value
        self.next=None
```

```python
class Linklist():
    def __init__(self):
        self.head=None
        self.n=0


    def insert_head(self,value):

        new_node=Node(value)
        new_node.next=self.head
        self.head=new_node
        self.n=self.n+1

    def __len__(self):
        return self.n
    
    def __str__(self):
        curr=self.head
        result=''
        while curr!=None:
            
            result=result+str(curr.data)+'->'
            curr=curr.next
        return result[:-2]
    
    def insert_tail(self,value):

        new_node=Node(value)
        curr=self.head

        while curr.next != None:
            curr=curr.next
        curr.next=new_node
        self.n+=1

    def insert_after(self,after,value):

        new_node=Node(value)

        curr=self.head

        while curr.next != None:
            if curr.data == after:
                new_node.next=curr.next
                curr.next=new_node
            curr = curr.next

        if curr.next==None:
            curr.next=new_node
        self.n=self.n+1

    def clear(self):
        self.n=0
        self.head=None

    def del_head(self):
        self.head=self.head.next
        self.n=self.n-1

    def del_tail(self):
        curr=self.head
        if curr.next==None:
            return self.del_head()
        while curr.next.next != None:
            curr=curr.next
        curr.next=None
        self.n=self.n-1

    def remove(self,value):
        curr=self.head
        if curr.data==value:
            return self.del_head()
        
        while curr.next !=None:
            if curr.next.data==value:
                curr.next=curr.next.next
                self.n=self.n-1
                return
            curr=curr.next
        
        if curr.data==value:
            return self.del_tail()
        
    def find(self,value):
        curr=self.head
        pos=0
        while curr!=None:
            if curr.data==value:
                return pos
            curr=curr.next
            pos=pos+1
        if curr==None:
            return "Index out of range"
        
    def __delitem__(self, key):
        curr=self.head
        pos=0
        while curr!=None:
            if pos==key:
                self.remove(curr.data)
                self.n=self.n-1
                return
            curr=curr.next
            pos+=1
        if curr==None:
            return "item not found"
        
    def __getitem__(self, key):
        curr=self.head
        pos=0
        while curr!=None:
            if pos==key:
                return curr.data
            pos=pos+1
            curr=curr.next
        if curr==None:
            return "index out of range"

        
    def replace_max(self,value):
        '''Replaces the max int with value'''

        curr=self.head
        max=curr

        while curr != None:
            if curr.data>max.data:
                max=curr
            curr=curr.next

        max.data=value

    def add_odd(self):
        '''return sum of all values whose index is add'''

        curr=self.head
        pos=0
        result=0

        while curr!=None:
            if pos%2!=0:
                result=result+curr.data
                
            curr=curr.next
            pos=pos+1
        return result
    

    def reverse(self):

        curr_node=self.head
        prev_node=None

        while curr_node != None:
            temp_node=curr_node.next
            curr_node.next=prev_node

            prev_node=curr_node
            curr_node=temp_node

        self.head=prev_node

    
    def change_sent(self):

        curr=self.head

        while curr != None:
            if curr.data=="*" or curr.data=="/":
                curr.data=' '
                if curr.next.data=="*" or curr.next.data=="/":
                    curr.next.next.data=curr.next.next.data.upper()
                    curr.next=curr.next.next
            
            curr=curr.next




```
